```python
import os
import torch
import nibabel
import numpy as np
from glob import glob

base_dir = "/home/yb107/cvpr2025/DukeDiffSeg/outputs/diffunet-binary-iterative/6.6/inference/_liver_inference"
all_pred_files = glob(os.path.join(base_dir, "*_pred.nii.gz"))
all_label_files = [f.replace("_pred.nii.gz", "_label.nii.gz") for f in all_pred_files]
all_pred_files, all_label_files
```

```python
def compute_volume(nifti_file):
    img = nibabel.load(nifti_file)
    data = img.get_fdata()
    voxel_volume = np.prod(img.header.get_zooms())  # in mm^3
    organ_volume = (
        np.sum(data > 0.5)
        * voxel_volume
        # np.sum(data > 0.5)
    )  # assuming binary mask with threshold 0.5
    return organ_volume


pred_volumes = [compute_volume(f) for f in all_pred_files]
label_volumes = [compute_volume(f) for f in all_label_files]
```
